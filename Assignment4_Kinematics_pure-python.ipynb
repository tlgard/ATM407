### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= (0 + C_1 + 0)+(0 + 0 + C_4) = C_1 + C_4$
* $ \zeta_{rel} = \frac{\partial }{\partial } + \frac{\partial }{\partial } = (-1)(0 + C_3 + 0)-(0 + 0 + C_2) = -C_3 + C_2 = (C_2 - C_3)$
* $ def1 = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = (0 + C_3 + 0)+(0 + 0 + C_2) = C_3 + C_2$
* $ def2 = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = (0 + C_1 + 0) - (0 + 0 + C_4) = C_1 - C_4$

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

## Answer to Exercise 4: 

Vorticity plus straight flow creates an area of cyclonic motion that varries with intensity and altitude. 

At lower levels, the flow moves in a cyclonic fashion, with a clearly defined area of weak winds.

At upper levels ... as the stregth of U increases, the are of weak winds shifts northwards. This causes the southerly winds to lose some of the rotation effects, becoming distinctively westerly. As the stregth continues to increase, the winds become almost entirely westerly with not signs of rotation. If we look at winds on a geopotential scale, winds at such high altitudes such as 200 mb or less are generally westerly and dont resemble any characteristics of a low pressure system below.
