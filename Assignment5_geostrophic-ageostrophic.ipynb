### Background and math

You will decompose the total wind into geostrophic and ageostrophic aspects: $\vec V = \vec V_g + \vec V_a$. 

* Horizontal acceleration equations (F=ma) for inviscid flow:
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* The geostrophic wind is defined as *the wind that would occur if the LHS were zero*: 

$$ 0 = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ 0 = - \frac{\partial \Phi}{\partial y} - fu_g $$

## Answer to exercise 3: Acceleration-related ageostrophic winds 

This area of strong geostrophic wind is associated with a ridge over the pacific northwest United States and Canada. It is further outlined by the geostrophic heights in blue. The geostrophic winds are also southwesterly, indicating a ridge, because the geostrophic and ageostrophic winds follow the same path. As the winds move to the top of the troph, they start to become northwesterly. This change in direction accounts for the acceleration. At the top of the ridge, a single frame would show westerly ageostrophic winds. Therefore, the coriolis force acting on the geostrophic winds is northerly, pointing 90° to the right of the ageostrophic wind. Therefore, the coriolis to the left of the troph is northwest and northeast to the right of the troph.
